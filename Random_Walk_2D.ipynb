```python
import numpy as np
import matplotlib.pyplot as plt
import sys
import math
import random
```

```python
nsteps = 1000
x_pos = sys.maxsize
init_pos = (0,0)
```

```python
def move_x(init_pos):
  movement = np.random.choice([-1,1])
  new_x = init_pos[0] + movement
  new_pos = (new_x, init_pos[1])
  return new_pos
```

```python
def move_y(init_pos):
  movement = np.random.choice([-1,1])
  new_y = init_pos[1] + movement
  new_pos = (init_pos[0], new_y)
  return new_pos
```

```python
def simulate(nsteps, init_pos):
  positions = []
  positions.append((0,0))
  displacements = []
  for i in range(nsteps):
    choice = np.random.choice([0,1])
    if choice == 0:
      new_pos = move_x(init_pos)
    else:
      new_pos = move_y(init_pos)
    displacements.append(np.linalg.norm(np.array(new_pos) - np.array((0,0))))
    positions.append(new_pos)
    init_pos = new_pos
  return positions, displacements
```

```python
positions, displacements = simulate(nsteps, init_pos)
```

```python
x_pos_history = []
y_pos_history = []
for i in range(len(positions)):
  x_pos_history.append(positions[i][0])
  y_pos_history.append(positions[i][1])

plt.scatter(x_pos_history[0], y_pos_history[0], color='red', label='Start')
plt.scatter(x_pos_history[-1], y_pos_history[-1], color='green', label='End')
plt.plot(x_pos_history, y_pos_history)
plt.xlabel("x")
plt.ylabel("y")
plt.title("2D Random Walk")
plt.legend()
plt.grid(True)
plt.show()
```

```python

plt.plot(displacements)
plt.xlabel("Time")
plt.ylabel("Displacement")
plt.title("Displacement vs Time")
plt.grid(True)
plt.show()
```

```python
displacements = np.array(displacements)
msd = np.cumsum(displacements**2)/np.arange(1,nsteps+1)
```

```python

plt.plot(msd)
plt.xlabel("Time")
plt.ylabel("MSD")
plt.title("MSD vs Time")
plt.grid(True)
plt.show()
```

```python

import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot(x_pos_history, y_pos_history, range(len(x_pos_history)))
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_zlabel("Time")
plt.title("3D Random Walk")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import sys
import math
import random
```

```python
nsteps = 1000
x_pos = sys.maxsize
init_pos = 0
```

```python
def move_x(init_pos):
  movement = np.random.choice([-1,1])
  new_x = init_pos + movement
  return new_x

```

```python
def simulate(nsteps, init_pos):
  x_pos_history = []
  displacments = []
  x_pos_history.append(init_pos)
  for i in range(nsteps):
    new_x = move_x(init_pos)
    x_pos_history.append(new_x)
    displacments.append(new_x - 0)
    init_pos = new_x
  return x_pos_history, displacments
```

```python
positions, displacements = simulate(nsteps, init_pos)

```

```python
plt.plot(positions)
plt.xlabel("nsteps")
plt.ylabel("positions")
plt.title("Random Walk")
plt.grid(True)
plt.show()
```

```python

plt.plot(displacements)
plt.xlabel("Time steps")
plt.ylabel("displacements")
plt.title("Random Walk - Displacements")
plt.grid(True)
plt.show()
```

```python
displacements = np.array(displacements)
msd = np.cumsum(displacements**2)/np.arange(1,nsteps+1)
```

```python

plt.plot(msd)
plt.xlabel("nsteps")
plt.ylabel("msd")
plt.title("Mean Squared Displacement")
plt.grid(True)
plt.show()
```

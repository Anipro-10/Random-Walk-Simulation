```python
import numpy as np
import matplotlib.pyplot as plt
import sys
import math
import random

#for 1D case without updating movement in edge case
nsteps = 1000
x_pos = 100 #(-100 to 100)
init_pos = 0

def move_x(init_pos):
  movement = np.random.choice([-1,1])
  if movement == 1 and init_pos == 100:
    return init_pos
  elif movement == -1 and init_pos == -100:
    return init_pos
  new_x = init_pos + movement
  return new_x

def simulate(nsteps, init_pos):
  x_pos_history = []
  displacments = []
  x_pos_history.append(init_pos)
  for i in range(nsteps):
    new_x = move_x(init_pos)
    x_pos_history.append(new_x)
    displacments.append(new_x - 0)
    init_pos = new_x
  return x_pos_history, displacments

positions, displacements = simulate(nsteps, init_pos)

plt.plot(displacements)
plt.xlabel("nsteps")
plt.ylabel("displacements")
plt.title("Random Walk - Displacements")
plt.grid(True)
plt.show()

displacements = np.array(displacements)
msd = np.cumsum(displacements**2)/np.arange(1,nsteps+1)

plt.plot(msd)
plt.xlabel("nsteps")
plt.ylabel("msd")
plt.title("Mean Squared Displacement")
plt.grid(True)
plt.show()

```

```python
#for 2D case without updating movement in edge case
nsteps = 200000
x_pos = 100
y_pos = 100
init_pos = (0,0)

def move_x(init_pos):
  movement = np.random.choice([-1,1])
  if movement == 1 and init_pos[0] == 100:
    return init_pos
  elif movement == -1 and init_pos[0] == -100:
    return init_pos
  new_x = init_pos[0] + movement
  new_pos = (new_x, init_pos[1])
  return new_pos

def move_y(init_pos):
  movement = np.random.choice([-1,1])
  if movement == 1 and init_pos[1] == 100:
    return init_pos
  elif movement == -1 and init_pos[1] == -100:
    return init_pos
  new_y = init_pos[1] + movement
  new_pos = (init_pos[0], new_y)
  return new_pos

def simulate(nsteps, init_pos):
  positions = []
  positions.append((0,0))
  displacements = []
  for i in range(nsteps):
    choice = np.random.choice([0,1])
    if choice == 0:
      new_pos = move_x(init_pos)
    else:
      new_pos = move_y(init_pos)
    displacements.append(np.linalg.norm(np.array(new_pos) - np.array((0,0))))
    positions.append(new_pos)
    init_pos = new_pos
  return positions, displacements

positions, displacements = simulate(nsteps, init_pos)

x_pos_history = []
y_pos_history = []
for i in range(len(positions)):
  x_pos_history.append(positions[i][0])
  y_pos_history.append(positions[i][1])

plt.scatter(x_pos_history[0], y_pos_history[0], color='red', label='Start')
plt.scatter(x_pos_history[-1], y_pos_history[-1], color='green', label='End')
plt.plot(x_pos_history, y_pos_history)
plt.xlabel("x")
plt.ylabel("y")
plt.title("2D Random Walk")
plt.legend()
plt.grid(True)
plt.show()

plt.plot(displacements)
plt.xlabel("Time")
plt.ylabel("Displacement")
plt.title("Displacement vs Time")
plt.grid(True)
plt.show()

displacements = np.array(displacements)
msd = np.cumsum(displacements**2)/np.arange(1,nsteps+1)

plt.plot(msd)
plt.xlabel("Time")
plt.ylabel("MSD")
plt.title("MSD vs Time")
plt.grid(True)
plt.show()

import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot(x_pos_history, y_pos_history, range(len(x_pos_history)))
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_zlabel("Time")
plt.title("2D Random Walk- xy vs time")
plt.show()
```

```python
#for 3D case without updating movement in edge case
nsteps = 6000
x_pos = 12
y_pos = 12
z_pos = 12
init_pos = (0,0,0)

def move_x(init_pos):
  movement = np.random.choice([-1,1])
  if movement == 1 and init_pos[0] == 12:
    return init_pos
  elif movement == -1 and init_pos[0] == -12:
    return init_pos
  new_x = init_pos[0] + movement
  new_pos = (new_x, init_pos[1], init_pos[2])
  return new_pos

def move_y(init_pos):
  movement = np.random.choice([-1,1])
  if movement == 1 and init_pos[1] == 12:
    return init_pos
  elif movement == -1 and init_pos[1] == -12:
    return init_pos
  new_y = init_pos[1] + movement
  new_pos = (init_pos[0], new_y, init_pos[2])
  return new_pos

def move_z(init_pos):
  movement = np.random.choice([-1,1])
  if movement == 1 and init_pos[2]==12:
    return init_pos
  elif movement == -1 and init_pos[2]==-12:
    return init_pos
  new_z = init_pos[2] + movement
  new_pos = (init_pos[0], init_pos[1], new_z)
  return new_pos

def simulate(nsteps, init_pos):
  positions = []
  positions.append((0,0,0))
  displacements = []
  for i in range(nsteps):
    choice = np.random.choice([0,1,2])
    if choice == 0:
      new_pos = move_x(init_pos)
    elif choice == 1:
      new_pos = move_y(init_pos)
    else:
      new_pos = move_z(init_pos)
    displacements.append(np.linalg.norm(np.array(new_pos) - np.array((0,0,0))))
    positions.append(new_pos)
    init_pos = new_pos
  return positions, displacements

positions, displacements = simulate(nsteps, init_pos)

plt.plot(range(len(displacements)), displacements)
plt.xlabel("Number of steps")
plt.ylabel("Displacement")
plt.title("Displacement vs. Number of Steps")
plt.grid(True)
plt.show()

displacements = np.array(displacements)
msd = np.cumsum(displacements**2)/np.arange(1,nsteps+1)

plt.plot(msd)
plt.xlabel("Time")
plt.ylabel("MSD")
plt.title("MSD vs Time")
plt.grid(True)
plt.show()

x_pos_history = []
y_pos_history = []
z_pos_history = []
for i in range(len(positions)):
  x_pos_history.append(positions[i][0])
  y_pos_history.append(positions[i][1])
  z_pos_history.append(positions[i][2])

import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.plot(x_pos_history, y_pos_history, z_pos_history)
ax.scatter(x_pos_history[0], y_pos_history[0], z_pos_history[0], color='green', label='Start')
ax.scatter(x_pos_history[-1], y_pos_history[-1], z_pos_history[-1], color='red', label='End')

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
plt.title('Random Walk in 3D')
plt.legend()
plt.show()

```
